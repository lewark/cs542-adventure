```python
import jericho
```

```python
GAMES_DIR = "z-machine-games-master/jericho-game-suite"
```

```python
env = jericho.FrotzEnv(f"{GAMES_DIR}/zork1.z5")
```

```python
# Warning: infinite loop
# Basic Usage Example from jericho docs.
initial_observation, info = env.reset()
done = False
while not done:
    # Take an action in the environment using the step fuction.
    # The resulting text-observation, reward, and game-over indicator is returned.
    observation, reward, done, info = env.step('open mailbox')
    # Total score and move-count are returned in the info dictionary
    print('Total Score', info['score'], 'Moves', info['moves'])
print('Scored', info['score'], 'out of', env.get_max_score())
```

```python
# Play the current game

print("Playing game. 'q' to quit")

initial_observation, info = env.reset()
print(initial_observation)
player_in = input("Answer prompt")
print('player:', player_in)

done = False
while not done and player_in != 'q':
    # Take an action in the environment using the step fuction.
    # The resulting text-observation, reward, and game-over indicator is returned.
    observation, reward, done, info = env.step(player_in)
    print(observation)
    print(f'\nWalkthrough: {env.get_walkthrough()}')
    player_in = input("Answer prompt")
    print('player:', player_in)
    # Total score and move-count are returned in the info dictionary
    if player_in == '$score':
        print('Total Score', info['score'], 'Moves', info['moves'])
        layer_in = input("Answer prompt")
        print('player:', player_in)
print('Scored', info['score'], 'out of', env.get_max_score())
```
